```python
import re
def replace_names(review, movie_title):
  review=review.lower()
  index=[]
  for match in re.finditer(movie_title, review):
    #print(review)
    #print(match.start(), match.end())
    start = match.start()
    end =match.end()
    index.append((start,end))
    # new_character = "Movie"
    # review = review[:start] + new_character + review[end:]
  start=index[0][0]
  end=index[0][1]
  movie_len=len(movie_title)
  # diff=0
  if movie_len>5:
    diff=movie_len-5
  else:
    diff=5-movie_len

  for i in range(0,len(index)):
      new_character = "Movie"
      review = review[:start] + new_character + review[end:]
      if i<len(index)-1:
        if movie_len>5:
          start=index[i+1][0]-diff
          end=index[i+1][1]-diff
        elif movie_len<5:
          start=index[i+1][0]+diff
          end=index[i+1][1]+diff
        else:
          start=index[i+1][0]
          end=index[i+1][1]
      else:
        pass
  return (review)
```

```python
def positive_score(token):
    pos=0
    tag=''
    synset=[]
    if token.pos_=="ADJ":
      tag='a'
    elif token.pos_=="ADV":
      tag='r'
    elif token.pos_=="NOUN":
      tag='n'
    elif token.pos_=="VERB":
      tag='v'
    else:
      pass
      #print(token.text,token.pos_,"nothing")
    text=token.text
    if tag!='':
      synset=wn.synsets(text, tag)
    else:
      pass
    if len(synset)>0:
        synset = synset[0]
        swn_synset = swn.senti_synset(synset.name())
        pos=pos+swn_synset.pos_score()
    else:
      pos=1.0
    if (pos==0.0):
      pos=pos+swn_synset.obj_score()
    else:
      pass
    return pos

def negative_score(token):
    neg=0
    tag=''
    synset=[]
    if token.pos_=="ADJ":
      tag='a'
    elif token.pos_=="ADV":
      tag='r'
    elif token.pos_=="NOUN":
      tag='n'
    elif token.pos_=="VERB":
      tag='v'
    else:
      #print(token.text,token.pos_,"nothing")
      pass
    text=token.text
    if tag!='':
      synset=wn.synsets(text, tag)
    else:
      pass
    if len(synset)>0:
        synset = synset[0]
        swn_synset = swn.senti_synset(synset.name())
        neg=neg+swn_synset.neg_score()
    else:
      neg=1.0
    if (neg==0.0):
      neg=neg+swn_synset.obj_score()
    else:
      pass
    return neg
```

```python
def check_aspect_term(aspect_term):
    Overall=["movie", "film","overall"]
    Cast=["act", "acting", "actress", "actor", "role", "portray",
          "character", "villain", "performance", "performed", "played", "casting", "cast",
          "character-development"]
    Director=["direct", "direction", "directing", "director", "filmed", "filming", "filmmaking",
              "filmmaker", "cinematic", "edition", "cinematography"]
    Story=["storyline", "story", "tale", "romance", "dialog", "script", "storyteller", "ending",
           "storytelling", "revenge", "betrayal", "plot", "writing", "twist", "drama", "humor", "entry", "concept",
           "moment", "sequence", "narrative", "suspense", "comedy", "fiction", "thriller",
           "science-fiction", "drama, sci-fi"]
    Scene=["scene", "scenery", "animation", "violence", "screenplay", "action", "special-effect",
           "stunt", "shot", "visual", "prop", "camera", "graphic", "visual-effect", "theme", "action-scene"]
    Music=["lyric", "sound", "music", "audio", "musical", "title-track", "sound-effect", "soundtrack", "song", "voice"]
    if (aspect_term in Overall ):
      return (("Overall",1))
    elif (aspect_term in Cast):
      return (("Cast",1))
    elif (aspect_term in Director):
      return (("Director",1))
    elif (aspect_term in Story):
      return (("Story",1))
    elif (aspect_term in Scene):
      return (("Scene",1))
    elif (aspect_term in Music):
      return (("Music",1))
    else:
       return(("nothing",-1))
```

```python
from collections import defaultdict
def feature_sentiment(reviews,pos1,neg1):
    opinion_words=[]
    res=""
    final_sent_lst=[]
    opinion_words=[item for item in pos1]
    opinion_words.extend(neg1)
    for rev in reviews:
      update = 0
      sent_dict = defaultdict(list)
      for movie in movie_names:
            if movie in rev:
              updated_review=replace_names(str(rev),movie)
              update=1
              break
            else:
              continue

      if (update==1):
        rev1=updated_review
      else:
        rev1=rev
      sentence = nlp(rev1)
      for token in sentence:

        if (token.pos_=="ADJ" or  token.pos_=="ADV"):
          if (token.lemma_  not in opinion_words) or (token.text not in opinion_words):
            res=find_pos__neg_word(token)
            if(res=='pos'):
              pos1.append(token.text)
              opinion_words.append(token.text)
            elif(res=='neg'):
              neg1.append(token.text)
              opinion_words.append(token.text)
          else:
            pass
        if token.text in pos1:
          score=positive_score(token)
        elif token.text in neg1:
          score=negative_score(token)
        else:
          pass
      for token in sentence:
          if (token.lemma_ in opinion_words) or (token.text in opinion_words):
              if (token.text in pos1) or (token.lemma_ in pos1):
                sentiment = 1
              else:
                sentiment = -1
              flag=find_hyphenated_word(token.text)
              if flag==1:
                if token.head.pos_=="NOUN":
                  if token.head.text!=token.text:
                    aspect_term=token.head.lemma_
                    (aspect_word,flag1)=check_aspect_term(aspect_term)
                    if flag1==1:
                      if sentiment==1 or sentiment==1.5:
                        senti="positive"
                      else:
                        senti="negative"
                      sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                    else:
                      pass
                  else:
                    pass
                else:
                  pass

              elif (token.dep_ == "advmod"):
                  continue
              elif (token.dep_ == "amod"):
                  if (token.head.text!=token.text):
                    aspect_term=token.head.lemma_
                    (aspect_word,flag1)=check_aspect_term(aspect_term)
                    if flag1==1:
                      if sentiment==1 or sentiment==1.5:
                        senti="positive"
                      else:
                        senti="negative"
                      sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                    else:
                      pass

                  else:
                    pass

              else:
                  for child in token.children:

                      if ((child.dep_ == "amod") or (child.dep_ == "advmod")) and (child.text in opinion_words):
                          sentiment *= 1.5
                      if child.dep_ == "neg":
                          sentiment *= -1
                  for child in token.children:

                      if (token.pos_ == "VERB") & (child.dep_ == "dobj"):
                          if(child.text!=token.text):
                              aspect_term=child.lemma_
                              (aspect_word,flag1)=check_aspect_term(aspect_term)
                              if flag1==1:
                                if sentiment==1 or sentiment==1.5:
                                  senti="positive"
                                else:
                                  senti="negative"
                                sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                              else:
                                pass
                          else:
                            pass
                          subchildren = []
                          conj = 0
                          for subchild in child.children:
                              if subchild.text == "and":
                                  conj=1
                              if (conj == 1) and (subchild.text != "and"):
                                  subchildren.append(subchild.text)
                                  conj = 0
                          for subchild in subchildren:
                              if(subchild!=token.text):
                                aspect_term=subchild
                                (aspect_word,flag1)=check_aspect_term(aspect_term)
                                if flag1==1:
                                  if sentiment==1 or sentiment==1.5:
                                    senti="positive"
                                  else:
                                    senti="negative"
                                  sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                                else:
                                  pass

                              else:
                                pass
                  # check for negation
                  for child in token.head.children:
                      noun = ""
                      if ((child.dep_ == "amod") or (child.dep_ == "advmod")) and (child.text in opinion_words):
                          sentiment *= 1.5
                      # check for negation words and flip the sign of sentiment
                      if (child.dep_ == "neg"):
                          sentiment *= -1



                  # check cconj and conj
                  for child in token.children:
                    noun=''
                    if (child.dep_=="conj"):
                      for grandchild in token.head.children:
                        if grandchild.dep_=='nsubj':
                          noun=grandchild.text
                        else:
                          pass

                      aspect_term=noun
                      (aspect_word,flag1)=check_aspect_term(aspect_term)
                      if flag1==1:
                        if sentiment==1 or sentiment==1.5:
                          senti="positive"
                        else:
                          senti="negative"

                        sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                        sent_dict[aspect_word].append((aspect_term,score,child.text,senti))
                      else:
                        pass

                    else:
                      pass
                  # check for nouns
                  for child in token.head.children:
                      noun = ""
                      if (child.pos_ == "NOUN") and (child.text not in sent_dict):
                          noun = child.text
                          # Check for compound nouns
                          for subchild in child.children:
                              if subchild.dep_ == "compound":
                                  noun = subchild.text + " " + noun
                          if (noun!=token.text):

                            aspect_term=noun
                            (aspect_word,flag1)=check_aspect_term(aspect_term)
                            if flag1==1:
                              if sentiment==1 or sentiment==1.5:
                                senti="positive"
                              else:
                                senti="negative"
                              sent_dict[aspect_word].append((aspect_term,score,token.text,senti))
                            else:
                              pass

                          else:
                            pass

      final_sent_lst.append(sent_dict)
    return (final_sent_lst)

pos1=list(positive_keywords)

neg1 = [x for x in negative_keywords if str(x) != 'nan']
(aspect_sentiment)=feature_sentiment(reviews,pos1,neg1)
```

```python
def emotion_checker(str1,sentiment,movie):
    emo_dict = te.get_emotion(str1)
    movie=movie.lower()
    #print(emo_dict)
    emotion_str =dict()
    val_list=list(emo_dict.values())
    key_list = list(emo_dict.keys())
    score=max(val_list)
    positions = []
    if sentiment=="Positive":
      happy_score=emo_dict["Happy"]
      surprise_score=emo_dict["Surprise"]
      if (happy_score==1):
        emotion_str[movie] = "Happy"
      elif (surprise_score==1):
        emotion_str[movie]  = "Surprise"
      elif (happy_score>surprise_score):
        emotion_str[movie]  = "Happy"
      elif (happy_score<surprise_score):
        emotion_str[movie]  = "Surprise"
      elif (happy_score==surprise_score):
        emotion_str[movie] ="Happy, Surprise"
      else:
        emotion_str[movie] =''

    elif sentiment=="Negative":
      angry_score=emo_dict["Angry"]
      sad_score=emo_dict["Sad"]
      if (angry_score==1):
        emotion_str[movie] = "Angry"
      elif (sad_score==1):
        emotion_str[movie] = "Sad"
      elif (angry_score>sad_score):
        emotion_str[movie] = "Angry"
      elif (angry_score<sad_score):
        emotion_str[movie] = "Sad"
      elif (angry_score==sad_score):
        emotion_str[movie]="Angry, Sad"
      else:
        emotion_str[movie]=''
    else:
      if score>=0.5:
          for i in range(len(val_list)):
              if val_list[i] == score:
                  positions.append(i)
              else:
                  continue
          emotions=[key_list[pos] for pos in positions]
          emotion_str[movie] = ','.join(emotions)
      else:
          emotion_str[movie]=''
    return emotion_str
#temp_movie_names=reviews=list(data1['Movie '])
emotion_strings=list(map(emotion_checker,reviews,overall_sentiment,movie_fst_col))
print(emotion_strings)
```
